```python
import cv2
from helper import *
%matplotlib inline

img_path = './images/iloveyouticket.jpg'
img_path = '/Users/cwq/code/text_renderer/output/default/00000094.jpg'
img = cv2.imread(img_path, 0)

def resize(img, scale=4, flag=cv2.INTER_LINEAR):
    resized_img = cv2.resize(img, None, fx=1/scale, fy=1/scale, interpolation=flag)
    resized_img = cv2.resize(resized_img, None, fx=scale, fy=scale, interpolation=flag)
    return resized_img

def show_grid(img, scales, flag=cv2.INTER_LINEAR, name=''):
    f, axarr = plt.subplots(1,len(scales)+1)

    f.set_figheight(25)
    f.set_figwidth(25)
    
    axarr[0].imshow(img, cmap='gray')
    for i, scale in enumerate(scales):
        axarr[i+1].imshow(resize(img, scale, flag), cmap='gray')

show_grid(img, [1.5, 2, 2.5, 3], flag=cv2.INTER_LINEAR)

show_grid(img, [1.5, 2, 2.5, 3], flag=cv2.INTER_CUBIC)

show_grid(img, [1.5, 2, 2.5, 3], flag=cv2.INTER_AREA)

show_grid(img, [1.5, 2, 2.5, 3], flag=cv2.INTER_NEAREST)

```
