# 目标检测常用指标
## 基本

- True positives : **正样本被正确识别为正样本。**
- True negatives: **负样本被正确识别为负样本。**
- False positives: **假的正样本，即负样本被错误识别为正样本。**
- False negatives: **假的负样本，即正样本被错误识别为负样本。**

## IoU

交并比，用来判断检测框和标记框的重合程度

## Accuracy

所有检测正确的样本（TP + TN）数除以总样本数

## Precision

TP/(TP+FP)，某一类的正确检测结果占该类总检测结果的比例，1为最佳，即没有认错（FP）

## Recall

TP/P，某一类正确检测的结果占左右正样本的比例，1为最佳，即找到了所有该类对象

## F-measure
$F=\frac{(a^2+1)Precision*Recall}{a^2(Precision+Recall)}$

当 $a=1$，即为 F1 值：

$F1=\frac{2Precision*Recall}{Precision+Recall}$

## Average Precision(AP)

PR 曲线下面的面积，通常来说一个越好的分类器，AP值越高

## mean Average Precision(mAP)

所有类别 AP 的均值

## ROC（Receiver Operating Characteristic）曲线

FPR：所有负样本（FP+TN）中错误的认为是正样本比例，即误报率

TPR：所有正样本（TP+FN）中检测正确的的概率，即命中率

以人脸识别为例子，如果调高相似度阈值，虽然 TPF 会，但是 FPR 也会增大，所以 FPR  越小，而且 TPR 越高越好。
