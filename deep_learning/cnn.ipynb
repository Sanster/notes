## AlexNet(2012)
2012 年 ImageNet 大赛的冠军，相比于以往没有使用 Deeping learning 的结果大幅降低了 top-5 错误率
![AlexNet](./images/AlexNet.png)

- 8 layers
- 卷积核：11x11, 5x5, 3x3
- max pooling
- dropout
- data augmentation
- ReLU
- SGD with momentum

## VGGNet(2014)
- 常用 VGG16/19
- 结构简单，只采用 3x3 的卷积核以及 2x2 的 pooling 层，有许多连续卷积层，参数量、计算量巨大。
![VGG16](./images/VGG16.png)

- 可以近似认为两个 3x3 卷积与单个 5x5 卷积的感受野范围相同，但参数更少
![3x3](./images/3x3.png)

## ResNet(2015)
一般会认为随着网络的深度增加，网络的性能也会越好，然而由于梯度消失和其它未知的原因，实际情况并不是如此。
残差网络添加了残差块，使得训练深达 152 层的神经网络成为可能，并能获得优秀的性能。

对于较浅的 resnet，均采用 3x3 的卷积核；对于较深的 resnet，会采用 1x1 的卷积减少参数，称这种结构为 bottleneck

![ResNetBlock](./images/ResNetBlock.jpeg)

残差结构可能被训练成近似单位映射，保证了该层的输入可以继续传递下去
